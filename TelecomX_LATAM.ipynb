```python
df_normalizado['account_Charges_Total'] = pd.to_numeric(df_normalizado['account_Charges_Total'], errors='coerce')
```

```python
for col in df_normalizado.select_dtypes(include='object').columns:
    print(f"{col}: {df_normalizado[col].unique()}\n")
```

```python
datos_normalizados = df_normalizado.drop_duplicates()

for col in ['account_Charges_Total', 'account_Charges_Monthly']:
    if col in datos_normalizados.columns:
        datos_normalizados[col] = pd.to_numeric(datos_normalizados[col], errors='coerce')
        datos_normalizados[col] = datos_normalizados[col].fillna(datos_normalizados[col].mean())

for col in datos_normalizados.select_dtypes(include='object').columns:
    datos_normalizados[col] = datos_normalizados[col].astype(str).str.strip().str.lower()

for col in datos_normalizados.select_dtypes(include='object').columns:
    datos_normalizados[col] = datos_normalizados[col].fillna('desconocido')

for col in datos_normalizados.select_dtypes(include='datetime64[ns]').columns:
    datos_normalizados[col] = datos_normalizados[col].dt.normalize()
```

```python
datos_normalizados.tail()
```

```python
datos_normalizados['Cuentas_Diarias'] = datos_normalizados['account_Charges_Monthly'] / 30
```

```python
datos_normalizados['Cuentas_Diarias']
```

```python
datos_normalizados.head()
```

```python

binario = {'yes': 1, 'no': 0, 'no phone service': 0, 'no internet service': 0}

columnas_binarias = [col for col in datos_normalizados.columns if datos_normalizados[col].nunique() <= 3 and datos_normalizados[col].dtype == 'object']

for col in columnas_binarias:
    datos_normalizados[col] = datos_normalizados[col].map(binario).fillna(datos_normalizados[col])

datos_normalizados = datos_normalizados.rename(columns={
    'customer_gender': 'Genero',
    'customer_SeniorCitizen': 'AdultoMayor',
    'customer_Partner': 'TienePareja',
    'customer_Dependents': 'TieneDependientes',
    'customer_tenure': 'MesesConEmpresa',
    'account_Contract': 'TipoContrato',
    'account_PaperlessBilling': 'FacturaDigital',
    'account_PaymentMethod': 'MetodoPago',
    'account_Charges_Monthly': 'FacturaMensual',
    'account_Charges_Total': 'FacturaTotal',
    'Churn': 'Cancelacion'
})
```
